```python
import pandas as pd

result_data = pd.read_csv('C:/Users/hyeseung/Desktop/4-2/ML/최종데이터/result_data.csv')
result_data
```

```python
태풍영향일자=pd.read_excel('C:/Users/hyeseung/Desktop/4-2/ML/최종데이터/태풍영향일자.xlsx')
태풍영향일자
```

```python
# 태풍영향일자 데이터를 리스트로 변환
태풍_dates = 태풍영향일자.values.flatten()
태풍_dates = [str(date).replace('"', '') for date in 태풍_dates if str(date) != 'nan']

# '수송일자'가 태풍영향일자에 포함되는지 확인하여 '태풍' 변수 생성
result_data['태풍'] = result_data['수송일자'].apply(lambda x: 1 if x in 태풍_dates else 0)

# 결과 확인
result_data[['수송일자', '태풍']].head(10)
```

```python
# Save the processed data to a new Excel file
file_path = "C:/Users/hyeseung/Desktop/4-2/ML/최종데이터/태풍추가_data.xlsx"
result_data.to_excel(file_path, index=False)

file_path
```

```python
data=pd.read_excel('C:/Users/hyeseung/Desktop/4-2/ML/subwaydatas/9.폭염.xlsx')
data
```

```python
data['날짜'] = data['날짜'].astype(str)

data['연도'] = data['연도'].astype(str)
data['월'] = data['월'].astype(str)
data['일'] = data['일'].astype(str)
```

```python
# '날짜' 칼럼에서 월과 일 추출
data['월'] = data['날짜'].str[:-2].str.pad(2, fillchar='0')
data['일'] = data['날짜'].str[-2:]

# '연도', '월', '일' 칼럼을 합쳐 'yyyy-mm-dd' 형태로 '연도날짜' 칼럼 생성
data['연도날짜'] = data['연도'] + '-' + data['월'] + '-' + data['일']

# '월'과 '일' 칼럼은 임시로 사용했으므로 삭제
data = data.drop(columns=['월', '일'])

# 변경된 데이터의 첫 5행 출력
data.head()
```

```python
# 폭염 데이터를 리스트로 변환
폭염_dates = data['연도날짜'].tolist()
폭염_dates
```

```python
# '수송일자'가 태풍영향일자에 포함되는지 확인하여 '태풍' 변수 생성
result_data['폭염'] = result_data['수송일자'].apply(lambda x: 1 if x in 폭염_dates else 0)

# 결과 확인
result_data[['수송일자', '폭염']].head(10)
```

```python
# Save the processed data to a new Excel file
file_path = "C:/Users/hyeseung/Desktop/4-2/ML/최종데이터/태풍폭염추가_data.xlsx"
result_data.to_excel(file_path, index=False)

file_path
```
